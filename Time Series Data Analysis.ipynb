## Working with Real Datasets

Datasets can be obtained from **Yahoo finance** or a package can be used : **quandl**.

## Using Quandl

```python
import quandl
data = quandl.get("WIKI/AAPL",start_date="1990-01-01",end_date="2018-12-31",collapse="weekly")
data = data[['Adj. Open','Adj. High','Adj. Low','Adj. Close','Adj. Volume']]
data.rename(columns={'Adj. Open':'Open','Adj. High':'High','Adj. Low':'Low','Adj. Close':'Close','Adj. Volume':'Volume'},inplace=True)

data.head(5)
```

```python
data.plot(y = ["Close"])
```

There is a sudden increase after 2010, frequent dips too, maybe due to market fluctuation. This is true for most Closing Price Time Series.

## Real Dataset

```python
import pandas as pd

df = pd.read_csv("USA_SP500_1985-2016_ordered.csv")
df.set_index("Date", inplace = True)
df.head()
```

```python
import matplotlib.pyplot as plt 

plt.figure(figsize=(20,6))
df['CD'].plot(label='CD')
df['CD.1'].plot(label='CD.1')
df['CD.2'].plot(label='CD.2')
plt.legend()
plt.title('Adjusted Closing Price')
plt.ylabel('Price ($)')
```
