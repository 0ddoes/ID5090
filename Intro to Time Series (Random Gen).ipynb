## Random Number Generation with Plotting

```python
import random
import matplotlib.pyplot as plt

ran = []
T = 1000
for i in range(T):
    ran.append(random.random())
    
plt.figure(figsize = (15,5))
plt.plot(ran)
plt.title("Random Number Time Series n = {} ".format(T))
plt.grid()
plt.xlim(0,T)
plt.show()
```

## Generate 10 Time Series Data

```python
def TS(T):
    ran = []
    for i in range(T):
        ran.append(random.random())
    return(ran)
    
TimeSeries = {}
T = 1000
for i in range(10):
    TimeSeries["TS{}".format(i+1)] = TS(T)
    
plt.figure(figsize = (15,5))
for i in range(10):
    plt.plot(TimeSeries["TS{}".format(i+1)], label = "TS{}".format(i+1))
plt.title("10 Random Number Time Series n = {} ".format(T))
plt.legend()
plt.grid()
plt.xlim(0,T)
plt.show()
```

## Convert Dictionary to a DataFrame

```python
import pandas as pd

rng = pd.date_range("2019-01-01", freq = 'MS', periods = T)
df = pd.DataFrame.from_dict(TimeSeries)
df = df.set_index(rng)  
df.head()
```

## Correlation Plot

```python
cor = df.corr()
display(cor)
```

```python
import seaborn as sns
plt.figure(figsize = (8,6))

sns.heatmap(cor, cmap = plt.cm.Oranges)
```

The correlation between these random values show that there isn't much relation between the values. Correlation in **most cases is 0**, and that is what is expected with **Random Data**. As length of Time Series increases, the Correlation tends to 0 more.
